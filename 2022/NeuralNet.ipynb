```python
import pandas as pd
import numpy as np

from keras.layers import Dense
from keras import Sequential
```

```python
data = pd.read_csv('Algorithm/mvp/mvp_data.csv')
features = ['VORP','WS','OWS','BPM','PER','PTS','FT','FTA','OBPM','FG','USG%','Seed','MP','GS']
```

```python
train_data = data[data['Year'] < 2022]
test_data  = data[data['Year'] == 2022]

X_train = np.array(train_data[features])
X_test =  np.array(test_data[features])

y_train = np.array(train_data['Share'])
y_test  = np.array(test_data['Share'])
```

```python
# Define the model architecture
model = Sequential([

    Dense(64, activation = 'relu', input_shape = (len(X_train[1]),)),
    Dense(32, activation = 'relu'),
    Dense(16, activation = 'relu'),
    Dense(1)]

    )
```

```python
# Compile the model
model.compile(loss      = 'mse',
              optimizer = 'adam',
              metrics   = ['mse', 'mae']
              )
```

```python
# Train the model
model.fit(X_train, y_train, epochs = 50, validation_split = 0.2)
```

```python
pred = model.predict(X_test)
```

```python
test_data['pred'] = pred
test_data.sort_values('pred', ascending = False).head(10)[['Player'] + features + ['Share', 'pred']]
```
